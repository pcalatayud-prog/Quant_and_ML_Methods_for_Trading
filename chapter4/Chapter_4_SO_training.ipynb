```python
#Looking for robust parameters
so_period = range(1,50,1)
so_d =  range(1,50,1)
```

```python
#Initialize empty Data Frame with index and columns determined
#accumulation = pd.DataFrame(0,index=long_opt, columns=short_opt)
best_parameters = pd.DataFrame(columns=["Performance","so_period","so_d","benchmark"])
```

```python
#Initialize empty Data Frame with index and columns determined
accumulation = pd.DataFrame(0,index=so_period, columns=so_d)
#Running the loop
for k in so_period:
    val = pd.DataFrame(0,index=so_period, columns=so_d)
    for j in so_d:
            #Set up Parameters
            standar_SO.set_parameters(k,j)
            #Runing Strategy
            output = standar_SO.test_strategy()
            #First argument is the row, the second one is the column
            val.loc[k,j]=float(output[0])
            if float(output[0])>1:
                list_row = [float(output[0]),k,j,float(output[1])]
                best_parameters.loc[len(best_parameters)] = list_row
            
    val=(val>1)
    val.astype(int)
    accumulation  = accumulation + val                
```

```python
maximus = best_parameters["Performance"].max()
best_parameters[best_parameters["Performance"]==maximus]
```

```python
accumulation.sum().sum()
```

```python
#Ploting heat Map
accumulation.style.background_gradient(cmap='Greens',axis=None,vmin=0.5)
#Not a very robust result. A lot variance respect to surrounding data
```
