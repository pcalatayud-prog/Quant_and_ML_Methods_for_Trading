```python
#Importing Required Libraries
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
import cufflinks as cf
from pandas_datareader import data
from pandas.tseries.frequencies import to_offset
```

```python
#Extracting S&P 500 data from Yahoo Finance
S_AND_P_500 = yf.download("^GSPC", start = "1995-01-01", end = "2022-06-01")
```

```python
S_AND_P_500
```

```python
#Backfill Missing Values
S_AND_P_500.bfill()
```

```python
#Checking NAN
S_AND_P_500.isna().sum().sum()
```

__SMA / EMA__

```python
qf = cf.QuantFig(df = S_AND_P_500.loc["2018":"2020"])
```

```python
qf.add_sma(periods = 50,color="red")
qf.add_ema(periods = 50)
```

```python
qf.iplot(title = "S&P 500. SMA(50) y EMA(50)", name = "S&P 500")
```
